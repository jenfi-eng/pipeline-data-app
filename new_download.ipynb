```python
def query_bank_data(check_application_data, i):
    transaction_data_query = f'''select *, fx_amount(convert_cents(operating_currency_balance_cents, operating_currency), operating_currency, 'USD', transaction_date) as USD_currency_balance,
    fx_amount(convert_cents(operating_currency_withdraw_cents, operating_currency), operating_currency, 'USD',transaction_date) as USD_currency_withdraw,
    fx_amount(convert_cents(operating_currency_deposit_cents, operating_currency), operating_currency, 'USD', transaction_date) as USD_currency_deposit,
    convert_cents(operating_currency_balance_cents, operating_currency) as operating_currency_balance,
    convert_cents(operating_currency_withdraw_cents, operating_currency) as operating_currency_withdraw,
    convert_cents(operating_currency_deposit_cents, operating_currency) as operating_currency_deposit,
    case when operating_currency_withdraw_cents > 0 then 1 else 0 end as debit_count,
    case when operating_currency_deposit_cents > 0 then 1 else 0 end as credit_count, {int(check_application_data['company_credit_application_id'][i])} as application_id
    from underwriting_bank_transactions
    where underwriting_bank_account_id = {int(check_application_data['id'][i])}
    and ((extract(year from transaction_date) = {check_application_data['lookback_year_1'][i]}
    and extract(month from transaction_date) = {check_application_data['lookback_month_1'][i]})
    or (extract(year from transaction_date) = {check_application_data['lookback_year_2'][i]}
    and extract(month from transaction_date) = {check_application_data['lookback_month_2'][i]})
    or (extract(year from transaction_date) = {check_application_data['lookback_year_3'][i]}
    and extract(month from transaction_date) = {check_application_data['lookback_month_3'][i]})
    or (extract(year from transaction_date) = {check_application_data['lookback_year_4'][i]}
    and extract(month from transaction_date) = {check_application_data['lookback_month_4'][i]})
    or (extract(year from transaction_date) = {check_application_data['lookback_year_5'][i]}
    and extract(month from transaction_date) = {check_application_data['lookback_month_5'][i]})
    or (extract(year from transaction_date) = {check_application_data['lookback_year_6'][i]}
    and extract(month from transaction_date) = {check_application_data['lookback_month_6'][i]}))
    '''
    return Jenfi.df_query(transaction_data_query, rebuild_cache=True)
```

```python
df = pd.DataFrame()
# for i in range(len(check_application_data)):
for i in range(len(check_application_data)):    
    import time
    start_at = time.perf_counter()
    transaction_data = query_bank_data(check_application_data, i)
    df = df.append(transaction_data)
    end_at = time.perf_counter()

    print(f"Query {i} took: {end_at - start_at}s.")
```

```python
import boto3
import sagemaker
from sagemaker.sklearn.model import SKLearnModel
```

```python
runtime = boto3.client("sagemaker-runtime")
```

```python
endpoint_name = "VnTwoMobFifteenPlus"
```

```python
from statsmodels.api import add_constant
input_predict = add_constant(feat_dataframe, has_constant='add')
input_predict
```

```python
input_predict_json = input_predict.to_json(orient='records', lines=True).encode("utf-8")
input_predict_json
```

```python
response = runtime.invoke_endpoint(
    EndpointName=endpoint_name,
    Body=input_predict_json,
    ContentType="application/json",
)
```

```python
result_json = response["Body"].read()
result_json
```

```python
import json
from pandas import json_normalize

y_pred = json.loads(result_json)
sagemaker_predicted_probability = y_pred['output'][0][1]
```
